```python
import tensorflow
from tensorflow import keras
from keras.datasets import mnist
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from keras.utils import to_categorical
```

```python
(train_data , train_labels) , (test_data, test_labels) = mnist.load_data()
train_data.shape
```

```python
train_data = train_data.reshape(train_data.shape + (1,))
test_data = test_data.reshape(test_data.shape  + (1,))
```

```python
train_data = train_data/255
test_data = test_data/255
```

```python
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
train_labels[0]
```

```python
datagen = ImageDataGenerator(
    vertical_flip=True,
    rotation_range=20,
    height_shift_range=0.2,
    width_shift_range=0,
    shear_range=0.2,
    fill_mode='nearest'
)
```

```python
i =0
for batch in datagen.flow(train_data, batch_size=1 ,
                          save_to_dir='digit_aug',
                          save_prefix='jpeg'):
    i+=1
    if(i>10):
        break
```

```python
datagen.fit(train_data)
```

```python
aug_gen = datagen.flow(train_data,train_labels,batch_size=32)
```

```python
for i in range(5):
    aug_img,label = aug_gen.next()
    plt.figure(figsize=(12,12))
    for j in range(32):
        plt.subplot(1,32,j+1)
        plt.imshow(aug_img[j].reshape(28,28), cmap='gray')
        plt.axis('off')
    plt.show()
```

```python
import tensorflow as tf
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(132 , activation="relu"),
    tf.keras.layers.Dense(10 , activation="softmax")
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_data,train_labels, epochs=10)
```

```python
model.fit(aug_gen, epochs=10)
```
